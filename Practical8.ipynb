```python
import pandas as pd
df = pd.read_csv("/Users/maheshshirsath/Downloads/Machine_learning/spotify_analysis_dataset.csv")

print("Dataset Loaded Successfully")
print(df.head())
```

```python
# Random Forest
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# -------------------------
# REQUIRED DEFINITIONS
# -------------------------
df['popular'] = (df['popularity'] >= 50).astype(int)

X_log = df[['danceability', 'energy']]
y_log = df['popular']

# -------------------------
# Train Random Forest
# -------------------------
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
rf.fit(X_log, y_log)

# -------------------------
# Prediction
# -------------------------
new_data = pd.DataFrame([[0.6, 0.75]], columns=['danceability', 'energy'])
prediction = rf.predict(new_data)
print("Random Forest Prediction:", prediction[0])

# -------------------------
# Decision Region Plot
# -------------------------
x_min, x_max = 0, 1
y_min, y_max = 0, 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

grid = pd.DataFrame(
    np.c_[xx.ravel(), yy.ravel()],
    columns=['danceability', 'energy']
)

Z = rf.predict(grid).reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
plt.scatter(
    X_log['danceability'],
    X_log['energy'],
    c=y_log,
    edgecolor='k',
    cmap='bwr',
    s=60
)

plt.scatter(
    new_data['danceability'],
    new_data['energy'],
    color='green',
    s=120,
    label='Prediction (0.6, 0.75)'
)

plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.title('Random Forest Decision Boundary')
plt.legend()
plt.show()
```
