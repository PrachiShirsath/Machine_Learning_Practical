```python
import pandas as pd
df = pd.read_csv("/Users/maheshshirsath/Downloads/Machine_learning/spotify_analysis_dataset.csv")

print("Dataset Loaded Successfully")
print(df.head())
```

```python
# Decision Tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# -------------------------
# REQUIRED DEFINITIONS
# -------------------------
# Convert popularity to binary
df['popular'] = (df['popularity'] >= 50).astype(int)

X_log = df[['danceability', 'energy']]
y_log = df['popular']

# -------------------------
# Train Decision Tree
# -------------------------
dt = DecisionTreeClassifier(random_state=42, max_depth=3)
dt.fit(X_log, y_log)

# -------------------------
# Prediction
# -------------------------
new_data = pd.DataFrame([[0.65, 0.8]], columns=['danceability', 'energy'])
prediction = dt.predict(new_data)
print("Decision Tree Prediction:", prediction[0])

# -------------------------
# Plot Tree
# -------------------------
plt.figure(figsize=(14, 8))
plot_tree(
    dt,
    feature_names=X_log.columns,
    class_names=['Not Popular', 'Popular'],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree: Popular vs Not Popular")
plt.show()
```
